The purpose here is to loop through all data files. In this case there is only one, but I chose to use this method should I choose to add more. The files will be read into a Pandas dataframe and then into a dictionary.

```python
import pandas
import numpy as np
files = ["options_KR.csv"]
data = {}
for f in files:
    df1 = pandas.read_csv(f)
    data[f.replace(".csv", "")] = df1
```
